```python
to_remove = ['Title', 'AddressLine1','AddressLine2','Suffix','City','LastUpdated','PhoneNumber','MiddleName','LastName']
df = df.drop(to_remove, axis=1)
```

```python
classification = {'discrete' : ['CustomerID','NumberCarsOwned','NumberChildrenAtHome','TotalChildren'],
'continous' : ['YearlyIncome'],
'nominal' : ['HomeOwnerFlag','FirstName','StateProvinceName','CountryRegionName','Gender','MaritalStatus','PostalCode'],
'ordinal' : ['Education'],
'interval' : ['BirthDate'],
'ratio' : ['YearlyIncome','NumberChildrenAtHome','TotalChildren']
}
```

```python
to_remove = ['CustomerID', 'FirstName', 'StateProvinceName', 'CountryRegionName', 'PostalCode']
df_new = df.drop(to_remove, axis=1)
```

```python
df_new.head()
```

```python
df_new.dtypes
```

```python
for col in df_new.columns:
  if df_new[col].dtype == "object":
    df_new[col] = df_new[col].fillna(df_new[col].mode()[0])
  else:
    df_new[col] = df_new[col].fillna(df_new[col].median())
```

```python
df_new["BirthDate"] = pd.to_datetime(df_new["BirthDate"], errors="coerce")

# Compute Age
reference_year = 2017
df_new["Age"] = reference_year - df_new["BirthDate"].dt.year

df_new[["BirthDate", "Age"]].head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_new[["Age_norm","Income_norm"]] = scaler.fit_transform(df_new[["Age","YearlyIncome"]])
```
